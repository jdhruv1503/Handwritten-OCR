```python
import pytesseract
from PIL import Image
import os
import numpy as np
from jiwer import wer, cer

# Function to process image with Tesseract
def get_text_from_image(image_path):
    try:
        img = Image.open(image_path)
        text = pytesseract.image_to_string(img).strip().upper()
        return text
    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")
        return ""

from tqdm import tqdm
# Process first 4000 images
results = []
for idx, row in tqdm(data.iterrows()):
    if idx >= 600:
        break
        
    filename = row['FILENAME']
    true_text = row['IDENTITY']
    
    # Get predicted text
    image_path = os.path.join('validation_v2/validation_filtered', filename)  # Adjust path as needed
    pred_text = get_text_from_image(image_path)
    
    results.append({
        'filename': filename,
        'true_text': true_text,
        'pred_text': pred_text
    })

# Convert results to DataFrame
```

```python
results
```

```python
results_df = pd.DataFrame(results)
```

```python
results_df['true_text'].values.tolist()
```

```python

word_error_rate = wer(results_df['true_text'].values.tolist(), results_df['pred_text'].values.tolist())
char_error_rate = cer(results_df['true_text'].values.tolist(), results_df['pred_text'].values.tolist())
```

```python
print(f"Word Error Rate: {word_error_rate:.4f}")
print(f"Character Error Rate: {char_error_rate:.4f}")
```

```python
exact_matches = (results_df['true_text'].str.replace(" ", "").str.upper() == results_df['pred_text'].str.replace(" ", "").str.upper())
```

```python
exact_matches
```

```python


# Calculate accuracy (exact matches)
exact_matches = (results_df['true_text'].str.replace(" ", "").str.upper() == results_df['pred_text'].str.replace(" ", "").str.upper()).mean()

print(f"Word Error Rate (WER): {word_error_rate:.4f}")
print(f"Character Error Rate (CER): {char_error_rate:.4f}")
print(f"Exact Match Accuracy: {exact_matches:.4f}")

# Display some example predictions
print("\nSample Predictions:")
print(results_df[['filename', 'true_text', 'pred_text']].head())
```
